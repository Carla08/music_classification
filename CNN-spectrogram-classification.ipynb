# Music classification CNN

## Extract features

```python
import os, sys
import numpy as np
import pandas as pd
import librosa
import IPython.display as ipd
from tqdm import tqdm
import matplotlib.pyplot as plt
import librosa.display
from collections import defaultdict
```

```python
# load raw audio files
genres_path = "Data/genres_original"

def load_audio_file(filename):
    return librosa.load(filename)

def load_label_with_audio_files():
    label_audio_dir = defaultdict(list)
    correct_loaded_files = 0
    for _dir in os.listdir(genres_path):
        genre = _dir
        genre_path = f"{genres_path}/{genre}"
        print(f"getting files from genre:{genre}")
        for filename in os.listdir(genre_path):
            try:
                print(f"getting audio from file: {filename}")
                audio_file = load_audio_file(f"{genre_path}/{filename}")
                label_audio_dir[genre].append(audio_file)
                correct_loaded_files += 1
            except Exception:
                print(f"Failed to load file: {filename}")
                continue
    return label_audio_dir, correct_loaded_files

files, num = load_label_with_audio_files()
print(f"Loaded {num}/1000")
```

```python

def get_spectogram_map(files_map):
    spectograms_map = defaultdict(list)
    for genre in files:
        print(f"creating spectograms for genre {genre}")
        for audiofile in files[genre]:
            #print(f"processing file {audiofile}")
            sound, sample_rate = audiofile
            X = librosa.stft(sound)
            Xdb = librosa.amplitude_to_db(abs(X))
            spectograms_map[genre].append(Xdb)
    return spectograms_map

spec_map = get_spectogram_map(files)
```

```python
def get_mfcc_map(files_map):
    mfcc_map = defaultdict(list)
    for genre in files:
        print(f"calculating mfcc for genre {genre}")
        for audiofile in files[genre]:
            sound, sample_rate = audiofile
            mfcc = librosa.feature.mfcc(y=sound, sr=sample_rate, n_mfcc= 10)
            mfcc_map[genre].append(mfcc)
    return mfcc_map

mfcc_map = get_mfcc_map(files)
```

```python
def get_mel_specs_map(files_map):
    mel_map = defaultdict(list)
    for genre in files:
        print(f"calculating mel spectograms for genre {genre}")
        for audiofile in files[genre]:
            sound, sample_rate = audiofile
            mel = librosa.feature.melspectrogram(y=sound)
            mel_db = librosa.power_to_db(mel)
            mel_map[genre].append(mel_db)
    return mel_map

mel_map = get_mel_specs_map(files)
```

```python
def transform_feature_map_to_np_float(feature_map):
    new_feat_map = defaultdict(list)
    for genre in feature_map:
        print(f"transforming to np float for genre {genre}")
        for feature in feature_map[genre]:
            new_feat_map[genre].append(feature.astype(np.float32))
    return new_feat_map

mel_map = transform_feature_map_to_np_float(mel_map)
mfcc_map = transform_feature_map_to_np_float(mfcc_map)
spec_map = transform_feature_map_to_np_float(spec_map)
```

## Create features and target splits

```python
# Create a map from genre to digits and viceversa
genre_digit_map = {}
digit_genre_map = {}

for i, genre in enumerate(files):
    genre_digit_map[genre] = i
    digit_genre_map[i] = genre
```

```python
def get_target_array():
    return [genre_digit_map[genre] for genre in files for file in files[genre]]

target = get_target_array()            
```

### Scale the data

### Spectogram scaling

```python
def flatten_specs_map(specs_map, _len = 1290):
    lst = []
    for genre in specs_map:
        for feature in specs_map[genre]:
            _feature = []
            for i_lst in feature: 
                _lst = i_lst[:_len]
                _feature.append(_lst)
            lst.append(_feature)
    return lst
```

```python
def get_specs_max(spec_map):
    maxes = []
    for i,genre in enumerate(spec_map):
        print(f"processing {i}/9...")
        for feature in spec_map[genre]:
            for lst in feature:
                maxes.append(max(lst))
    return max(maxes)
    
specs_max = get_specs_max(spec_map)
```

```python
flattened_specs = flatten_specs_map(spec_map)
scaled_specs = flattened_specs/specs_max
```

### Scale MFCC

```python
mfcc_max = get_specs_max(mfcc_map)
```

```python
flattened_mfcc = flatten_specs_map(mfcc_map)
scaled_mfcc = flattened_specs/mfcc_max
```

### Scale MEL spectograms

```python
mel_max = get_specs_max(mel_map)
mel_max
```

```python
flattened_mel = flatten_specs_map(mel_map)
scaled_mel = flattened_specs/mel_max
```

```python
from sklearn.model_selection import train_test_split
# split train-test data
spec_train, spec_test, mfcc_train, mfcc_test, mel_train, mel_test, y_train, y_test = train_test_split(scaled_specs, scaled_mfcc, scaled_mel, target, test_size= 0.2)
```

```python
y_train = tensorflow.keras.utils.to_categorical(np.array(y_train))
print(len(y_train))
```

```python
model = Sequential()
model.add(Conv2D(8, (3,3), activation= 'relu', input_shape= spec_train[0].shape, padding= 'same'))
model.add(MaxPooling2D((4,4), padding= 'same'))
model.add(Conv2D(16, (3,3), activation= 'relu', padding= 'same'))
model.add(MaxPooling2D((4,4), padding= 'same'))
model.add(Conv2D(32, (3,3), activation= 'relu', padding= 'same'))
model.add(MaxPooling2D((4,4), padding= 'same'))
model.add(Conv2D(64, (3,3), activation= 'relu', padding= 'same'))
model.add(MaxPooling2D((4,4), padding= 'same'))
model.add(Conv2D(64, (3,3), activation= 'relu', padding= 'same'))
model.add(MaxPooling2D((4,4), padding= 'same'))
model.add(Flatten())
model.add(Dense(128, activation= 'relu'))
model.add(Dense(64, activation= 'relu'))
model.add(Dense(10, activation= 'softmax'))

model.compile(optimizer= 'Adam', loss= 'categorical_crossentropy')

model.summary()

# Train Model 1

checkpoint = ModelCheckpoint(os.getcwd()+"/models/new_spec_model_spectrogram1_{epoch:03d}.h5", period= 5)

model.fit(spec_train, y_train, epochs= 50, callbacks= [checkpoint], batch_size= 32, verbose= 1)
#model.save(os.getcwd() + "/models/new_spec_model_spectrogram1.h5")
```
